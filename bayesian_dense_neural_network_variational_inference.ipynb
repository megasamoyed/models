```python
%matplotlib inline
import theano
import theano.tensor as T
floatX = theano.config.floatX

import pymc3 as pm
from pymc3.variational.callbacks import CheckParametersConvergence
import arviz as az

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings

from sklearn.preprocessing import scale
from sklearn.model_selection import train_test_split

filterwarnings('ignore')
sns.set_style('white')
```

# Get Data

```python
data = np.load("sample_data/fgs.npy" , allow_pickle = True).item()

# Combine Seperate DFs into one
ret_df = pd.DataFrame()
for k , v in data.items():
    # Create Returns and Directional Movement
    v.sort_index(ascending = True, inplace = True)
    v["ret"] = v["close"].pct_change()
    v.dropna(axis = 0 , inplace = True)
    v["dir"] = v["ret"].apply(lambda x: 1 if x >= 0  else 0)
    v["prod"] = k
    ret_df = pd.concat([ret_df , v[["ret"]]] , axis = 1)
    
ret_df.columns = data.keys()
```

```python
x_values = ret_df[["F","G"]].values
y_values = ret_df["S"].values
x_values = x_values.astype(floatX)
y_values = y_values.astype(floatX)
x_train, x_test, y_train, y_test = train_test_split(x_values, y_values , test_size=.5)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_train[:,0] , y_train , color='g')
ax.scatter(x_train[:,1] , y_train , color='r')
```

# Build Model

```python
class BayesianDNN():
    
    def __init__(self , x_train, y_train):
        
        self.x_train = x_train
        self.y_train = y_train
        self.input_shape = self.x_train.shape[1]
        
    
    def construct(self):
        # Defaults with 2 layer
        self.n_hidden = 5
        init_1 = np.random.randn(self.input_shape,  self.n_hidden).astype(floatX)
        init_2 = np.random.randn( self.n_hidden,  self.n_hidden).astype(floatX)
        init_out = np.random.randn( self.n_hidden ).astype(floatX) 
        
        with pm.Model() as self.neural_network:
            
            # Layers
            ann_input = pm.Data('ann_input', self.x_train)
            weights_in_1 = pm.Normal('w_in_1', 0, sigma=1 , shape=(self.input_shape, self.n_hidden), testval=init_1)
            act_1 = pm.math.tanh(pm.math.dot(ann_input, weights_in_1))
            
            weights_1_2 = pm.Normal('w_1_2', 0 , sigma=1, shape=(self.n_hidden, self.n_hidden), testval=init_2)
            act_2 = pm.math.tanh(pm.math.dot(act_1, weights_1_2))
            
            weights_2_out = pm.Normal('w_2_out', 0, sigma=1, shape=(self.n_hidden,), testval=init_out)
            act_out = pm.math.tanh(pm.math.dot(act_2, weights_2_out))
           
            ann_output = pm.Data('ann_output', self.y_train)
            
            # student params
            sigma = pm.HalfNormal('sigma', sd=10)
#             dof = pm.Exponential('dof', 1/30)
#             y = pm.StudentT('y', mu = act_out, sd=sigma, nu=dof, observed=ann_output)
            y = pm.Normal('y', mu = act_out, sd=sigma , observed=ann_output)
            
    
    def fit(self):
        with self.neural_network:
            self.inference = pm.ADVI()
            self.approx = pm.fit(n=30000, method=self.inference )
```

# Fit Model

```python
dnn = BayesianDNN(x_train, y_train)
dnn.construct()
dnn.fit()
```

```python
plt.plot(-dnn.inference.hist, label='new ADVI', alpha=.3)
plt.plot(dnn.approx.hist, label='old ADVI', alpha=.3)
plt.legend()
plt.ylabel('ELBO')
plt.xlabel('iteration');
```

# Evaluate

```python
trace = dnn.approx.sample(draws=5000)
pm.set_data(new_data={'ann_input': x_test, 'ann_output': y_test}, model = dnn.neural_network)
y_pred = pm.sample_posterior_predictive(trace, samples=500, progressbar = False, model = dnn.neural_network)
data_ppc = az.from_pymc3(trace=trace, posterior_predictive=y_pred)
ax = az.plot_ppc(data_ppc, figsize=(12, 6), mean=False)
```
